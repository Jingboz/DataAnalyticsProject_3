```python
# Dependencies
import matplotlib.pyplot as plt
import requests
import pandas as pd
from config import api_key
from pprint import pprint
```

```python
# Save config information.
url = "https://demo.openweathermap.org/energy/1.0/solar/data?"

# Build partial query URL
query_url = f"{url}appid={api_key}&date=2023-03-30"
```

```python

coordi = [["Perth",31.9523,115.8613],["Karratha",20.7337,116.8447]]
# set up lists to hold reponse info
city_solar = []

# Loop through the list of cities and perform a request for data on each
for city in coordi:
    response = requests.get(f"{query_url}&lat={city[1]}+&lon=+{city[2]}").json()
    response["city"] = city[0]
    city_solar.append(response)
pprint(city_solar)

```

```python
import json
```

```python

with open("./output/output.json", "w") as outfile:
    json.dump(city_solar, outfile)
```
